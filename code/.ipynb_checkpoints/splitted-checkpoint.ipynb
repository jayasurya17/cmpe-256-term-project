```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('../dataset/clothing_data_processed.csv')
```

```python

clothing_train, clothing_test = train_test_split(df, test_size=.2, random_state=1017)

clothing_train.to_csv('../dataset/train.csv', index = False)
clothing_test.to_csv('../dataset/test.csv', index = False)
```
